## Tuples
Les tuples sont les cousins immuables des listes. Presque tout ce que vous pouvez faire sur une liste sans la modifier, vous pouvez le faire sur un tuple. Vous spécifiez un tuple en utilisant des parenthèses (ou rien) au lieu de crochets :

```python
my_list = [1, 2]
my_tuple = (1, 2)
other_tuple = 3, 4
my_list[1] = 3      # ma_liste est maintenant [1, 3]

try:
    my_tuple[1] = 3
except TypeError:
    print("impossible de modifier un tuple")
```

Les tuples sont un moyen pratique de renvoyer plusieurs valeurs à partir de fonctions :

```python
def sum_and_product(x, y):
    return x+y, x*y

sp = sum_and_product(2, 3)    # est égal à (5, 6)
s, p = sum_and_product(5, 10) # s vaut 15, p vaut 50
```

## Dictionnaires

```python
empty_dict = {}                       # Pythonique
empty_dict2 = dict()                  # moins pythonique
grades = { "Alice" : 80, "Bob" : 95 } # dictionnaire littéral
```

### Compteur
Un **Compteur** transforme une séquence (liste ou tuple) de valeurs en un objet de type dict mappant les clés aux comptes. Il peut être pratique de créer des histogrammes par exemple :

```python
from collections import Counter
c = Counter([0, 1, 2, 0]) # c est (essentiellement) { 0 : 2, 1 : 1, 2 : 1 }
```

Cela nous donne un moyen très simple de compter les éléments distincts dans un document :

```python
document=['a','b','b','b','c','c','c','c','c','c','c','c','c','c']
word_counts = Counter(document) # est {'a' : 1, 'b' : 3, 'c' : 10})
```

```python
word_counts
```

Une instance **Counter** possède une méthode most_common qui est fréquemment utile :

```python
# imprimez les 10 mots les plus courants et leur nombre
for word, count in word_counts.most_common(10):
    print(word, count)
```

## Le hasard
Au fur et à mesure que nous apprenons la science des données, nous aurons fréquemment besoin de générer des nombres aléatoires, ce que nous pouvons faire avec le module random :

```python
import random
four_uniform_randoms = [random.random() for _ in range(4)]

# [0.8444218515250481, # random.random()
# 0,7579544029403025, # produit des nombres
# 0,420571580830845, # uniformément entre
# 0.25891675029296335] #0 et 1
```

Le module random produit en fait des nombres pseudo-aléatoires (c'est-à-dire déterministes) basés sur un état interne que vous pouvez définir avec **random.seed** si vous souhaitez obtenir des résultats reproductibles :

```python
random.seed(10)         # mettre la graine à 10
print(random.random())  # 0.57140259469
random.seed(10)         # remettre la graine à 10
print(random.random())  # 0,57140259469 encore
```

La fonction **random.randrange** prend 1 ou 2 arguments et renvoie un élément choisi aléatoirement dans la **range()** correspondante :

```python
random.randrange(10)   # choisir au hasard dans la plage (10) = [0, 1, ..., 9]
random.randrange(3, 6) # choisir au hasard dans la plage (3, 6) = [3, 4, 5]
```

La fonction **random.shuffle** réorganise aléatoirement les éléments d'une liste :

```python
up_to_ten = [x for x in range(10)]
random.shuffle(up_to_ten)
print(up_to_ten)
# [6, 7, 4, 3, 5, 2, 0, 1, 9, 8]
# (vos résultats seront probablement différents)
```

Si vous devez choisir au hasard un élément dans une liste, vous pouvez utiliser **random.choice** :

```python
my_best_friend = random.choice(["Alice", "Bob", "Charlie"])
# "Alice" pour moi
```

Et si vous devez choisir au hasard un échantillon d'éléments sans remplacement (c'est-à-dire sans doublons), vous pouvez utiliser **random.sample** :

```python
lottery_numbers = [x for x in range(60)]
winning_numbers = random.sample(lottery_numbers, 6)
# [16, 36, 10, 6, 25, 9]
```

Pour choisir un échantillon d'éléments avec remplacement (c'est-à-dire autoriser les doublons), vous pouvez simplement effectuer plusieurs appels à **random.choice** :

```python
four_with_replacement = [random.choice(range(10))
                         for _ in range(4)]
# [9, 4, 4, 2]
```

Étant donné une instance de **Set**, nous devrons pouvoir y ajouter des éléments, en supprimer des éléments et vérifier si elle contient une certaine valeur. Nous allons créer tout cela en tant que fonctions membres, ce qui signifie que nous y accéderons avec un point après un objet **Set** :

```python
# par convention, on donne aux classes des noms PascalCase
class Set:
    # ce sont les fonctions membres
    # chacun prend un premier paramètre "self" (autre convention)
    # qui fait référence à l'objet Set particulier utilisé

    def __init__(self, values=None):
        """This is the constructor.
        It gets called when you create a new Set. You would use it like
        s1 = Set()          # ensemble vide
        s2 = Set([1,2,2,3]) # initialiser avec les valeurs """
        self.dict = {}  # chaque instance de Set a sa propre propriété dict
                        # c'est ce que nous utiliserons pour suivre les adhésions
        if values is not None:
            for value in values:
                self.add(value)
    def __repr__(self):
        """this is the string representation of a Set object
        if you type it at the Python prompt or pass it to str()"""
        return "Set: " + str(self.dict.keys())

    # nous représenterons l'adhésion en étant une clé dans self.dict avec une valeur True
    def add(self, value):
        self.dict[value] = True

    # la valeur est dans l'ensemble si c'est une clé dans le dictionnaire
    def contains(self, value):
        return value in self.dict

    def remove(self, value):
        del self.dict[value]
```

Que l'on pourrait alors utiliser comme :

```python
s = Set([1,2,3])
s.add(4)
print(s.contains(4))  # Vrai
s.remove(3)
print(s.contains(3))  # FAUX
```

## Énumérer
Fréquemment, vous souhaiterez parcourir une liste et utiliser à la fois ses éléments et leurs index :

```python
# pas pythonique
for i in range(len(documents)):
    document = documents[i] do_something(i, document)

# pas non plus Pythonic
i=0
for document in documents:
    do_something(i, document)
    i+=1
```

La solution Pythonic est **enumerate**, qui produit des tuples (index, élément) :

```python
for i, document in enumerate(documents):
    do_something(i, document)
```

De même, si nous voulons juste les index :

```python
for i in range(len(documents)): do_something(i)   # pas pythonique
for i, _ in enumerate(documents): do_something(i) # Pythonique
```

## zip et déballage des arguments
Nous devrons souvent compresser deux ou plusieurs listes ensemble. **zip** transforme plusieurs listes en une seule liste de tuples d'éléments correspondants :

```python
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
zip(list1, list2) # est [('a', 1), ('b', 2), ('c', 3)]
```

Si les listes sont de longueurs différentes, **zip** s'arrête dès la fin de la première liste. Vous pouvez également « décompresser » une liste en utilisant une astuce étrange :

```python
pairs = [('a', 1), ('b', 2), ('c', 3)]
letters, numbers = zip(*pairs)
```

L'astérisque effectue le **déballage des arguments**, qui utilise les éléments des paires comme arguments individuels pour compresser. Cela se termine comme si vous aviez appelé :

zip(('a', 1), ('b', 2), ('c', 3)) # qui renvoie [('a','b','c'), ('1','2','3')]

Vous pouvez utiliser le déballage des arguments avec n'importe quelle fonction :

```python
def add(a, b):
    return a + b
add(1, 2)    # renvoie 3
add([1, 2])  # TapezErreur !
add(*[1, 2]) # renvoie 3
```

## Graphiques à barres
Un graphique à barres est un bon choix lorsque vous souhaitez montrer comment une quantité varie parmi un ensemble discret d'éléments.
Par exemple, l’intrigue suivante montre combien d’Oscars ont été remportés par chacun des différents films :

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca",
          "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# les barres ont par défaut une largeur de 0,8,
# nous ajouterons donc 0,1 aux coordonnées de gauche
# pour que chaque barre soit centrée
xs = [i + 0.1 for i, _ in enumerate(movies)]

# barres de tracé avec coordonnées x gauche [xs], hauteurs [num_oscars]
plt.bar(xs, num_oscars)
plt.ylabel("# des Oscars")
plt.title("My Favorite Movies")

# étiqueter l'axe X avec les noms de films dans les centres des bars
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

Un graphique à barres peut également être un bon choix pour tracer des histogrammes de valeurs numériques regroupées, afin d'explorer visuellement la manière dont les valeurs sont distribuées, comme dans le tracé suivant :

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10

histogram = Counter(decile(grade) for grade in grades)
plt.bar(# décaler chaque barre vers la gauche de 4
        [x - 4 for x in histogram.keys()],
        # donnez à chaque barre sa bonne hauteur
        histogram.values(),
        # donner à chaque barre une largeur de 8
        8)

plt.axis([-5, 105, 0, 5])   # axe x de -5 à 105,
                            # axe y de 0 à 5

plt.xticks([10 * i for i in range(11)]) # Étiquettes de l'axe X à 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# des étudiants")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

Le troisième argument de **plt.bar** spécifie la largeur de la barre.
Ici nous avons choisi une largeur de 8 (ce qui laisse un petit écart entre les barres, puisque nos godets ont une largeur de 10).
Et nous avons décalé la barre de 4 vers la gauche, de sorte que (par exemple) la barre « 80 » ait ses côtés gauche et droit à 76 et 84, et (donc) son centre à 80.
L'appel à **plt.axis** indique que nous voulons que l'axe des x soit compris entre -5 et 105 (afin que les barres « 0 » et « 100 » soient entièrement affichées) et que l'axe des y doit être compris entre 0 et 5.
Et l'appel à **plt.xticks** place les étiquettes de l'axe des x à 0, 10, 20, ..., 100.
Soyez judicieux lorsque vous utilisez plt.axis().

## Graphiques linéaires
Comme nous l'avons déjà vu, nous pouvons créer des graphiques linéaires en utilisant **plt.plot()**.
Il s'agit d'un bon choix pour montrer les tendances, comme l'illustre cette figure :

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# nous pouvons passer plusieurs appels à plt.plot
# pour afficher plusieurs séries sur le même graphique
plt.plot(xs, variance, 'g-', label='variance')      # ligne continue verte
plt.plot(xs, bias_squared, 'r-.', label='bias^2')   # ligne pointillée rouge
plt.plot(xs, total_error, 'b:', label='total error')# ligne pointillée bleue

# parce que nous avons attribué des étiquettes à chaque série
# nous pouvons obtenir une légende gratuitement
# loc=9 signifie "en haut au centre"
plt.legend()
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

## Nuages de points
Un nuage de points est le bon choix pour visualiser la relation entre deux ensembles de données appariés.
Par exemple, la figure ci-dessus illustre la relation entre le nombre d’amis d’un utilisateur et le nombre de minutes qu’il passe en ligne chaque jour :

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
# étiquetez chaque point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 # mettre l'étiquette avec sa pointe
                 xy=(friend_count, minute_count),
                 # mais légèrement décalé
                 xytext=(5, -5),
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# d'amis")
plt.ylabel("daily minutes spent online")
plt.show()
```

Si vous dispersez des variables comparables, vous pourriez obtenir une image trompeuse si vous laissez matplotlib choisir l'échelle, comme dans cette figure :

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

Si nous incluons un appel à **plt.axis("equal")**, le tracé montre plus précisément que la plupart des variations se produisent lors du test 2 :

```python
plt.scatter(test_1_grades, test_2_grades)
plt.axis("equal")
plt.title("Axes Are Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```
